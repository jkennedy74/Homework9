```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
w_df = pd.read_csv("weatherpy.csv")
```

```python
w_df.head()
```

```python
# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Prints: [8.0, 6.0]
print ("Current size:", fig_size)
 
# Set figure width to 12 and height to 9
fig_size[0] = 20
fig_size[1] = 10
plt.rcParams["figure.figsize"] = fig_size
```

```python
plt.scatter(x=w_df['Latitude'], y=w_df['Max Temp'],
            c = w_df['Max Temp'], cmap='plasma', alpha= .80, s = 200, linewidth=2, edgecolors='#3f3f3f')
ax = plt.gca()

plt.colorbar(label="MAX Temp")
plt.xlabel("Latitude")
plt.ylabel("Temparature (F)")
plt.ylim(-10, 110)
plt.xlim(-95, 95)
plt.title("Max Temperature vs Latitude")

plt.savefig('Temp')

plt.show()
```

```python
plt.scatter(x=w_df['Latitude'], y=w_df['Humidity'],
            c = w_df['Humidity'], cmap='plasma_r', alpha= .80, s = 200, linewidth=2, edgecolors='#3f3f3f')
ax = plt.gca()

plt.colorbar(label="% Humidity")
plt.xlabel("Latitude")
plt.ylabel("% Humidity")
plt.ylim(0,100)
plt.xlim(-95, 95)
plt.title("Humidity vs Latitude")

plt.savefig('Humidity')

plt.show()
```

```python
plt.scatter(x=w_df['Latitude'], y=w_df['Cloudiness'],
            c = w_df['Cloudiness'], cmap='plasma_r', alpha= .80, s = 200, linewidth=2, edgecolors='#3f3f3f')
ax = plt.gca()

plt.colorbar(label="% Cloudiness")
plt.xlabel("Latitude")
plt.ylabel("% Cloudiness")
plt.ylim(0, 100)
plt.xlim(-95, 95)
plt.title("Cloudiness vs Latitude")

plt.savefig('Cloudiness')

plt.show()
```

```python
plt.scatter(x=w_df['Latitude'], y=w_df['Wind Speed'],
            c = w_df['Wind Speed'], cmap='plasma_r', alpha= .80, s = 200, linewidth=2, edgecolors='#3f3f3f')
ax = plt.gca()

plt.colorbar(label="Wind Speed (mph)")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed")
plt.title("Wind Speed vs Latitude")
plt.ylim(0, 50)
plt.xlim(-95, 95)

plt.savefig('WindSpeed')

plt.show()
```
